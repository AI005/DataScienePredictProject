```python
import pandas as pd
import re
```

```python
df = pd.read_csv('tgdđdata_full.csv',index_col=0)
df.head(20)
```

```python
df.isnull().sum()
```

```python
columns = ["Company","CPU", "RAM", "Bus", "SSD", "HDD", "GPU_name", "GPU_ram", "Screen", "Độ phân giải","Weight" ,"Price"]
new_df = pd.DataFrame([], columns=columns)
new_df
```

```python
new_df['Company'] = df['Name'].apply(lambda x: x.split(' ')[1])
```

```python
new_df['Company']
```

```python
df['CPU'].unique()
```

```python
new_df['CPU'] = df['CPU'].apply(lambda x: " ".join((x.split(' ')[0:3] + x.split(' ')[5:6])).replace(",", "") if "Intel Core" in x else " ".join(x.split(' ')[0: 4]).replace(",", ""))
new_df['CPU'] = new_df['CPU'].apply(lambda x: re.sub('\d\.\d+', '', x))
new_df['CPU'] = new_df['CPU'].apply(lambda x: x[:x.rfind(" ")] + "-" + x[x.rfind(" ")+1:] if "Intel Core" in x else x)
new_df['CPU'] = new_df['CPU'].apply(lambda x: x[:x.rfind(" ")] + "-" + x[x.rfind(" ")+1:] if "AMD" in x else x)
new_df['CPU']
```

```python
new_df['RAM'] = df['RAM'].apply(lambda x: re.findall(r'([^,]+)',x)[0])
new_df['RAM']
```

```python
new_df['Bus'] = df['RAM'].apply(lambda x: x[x.rfind(",")+1:])
new_df['Bus']
```

```python
def find_ssd(x):
    cases_SSD = {'SSD 512':512, 'SSD: 512': 512, 'SSD 256': 256, 'SSD: 128': 128, 'SSD: 256': 256, 'SSD 128': 128}
    
    for key, value in cases_SSD.items():
        if key in x:
            return value
        
    return 0

new_df['SSD'] = df['Ổ cứng'].apply(find_ssd)
new_df['SSD'].unique()
```

```python
new_df['HDD'] = df['Ổ cứng'].apply(lambda x: float(re.findall(r'(\d+)',x)[0]) if x.startswith('HDD') else 0)
new_df['HDD'] = new_df['HDD'].replace(1, 1000).astype(float)
new_df['HDD']
```

```python
new_df['Độ phân giải'] = df['Màn hình'].apply(lambda x: re.findall(r'\(([^\)]+)\)', x)[0])
new_df['Độ phân giải']
```

```python
new_df["Price"] = df["Price"].apply(lambda x: float("".join(re.findall(r'(\d+)', x[:-1]))))
new_df["Price"]
```

```python
new_df['Screen'] = df['Màn hình'].apply(lambda x: re.findall(r'(^\d+.\d+|\d+)', x)[0])
new_df['Screen']
```

```python
new_df['Weight'] = df['Kích thước'].apply(lambda x: re.findall(r'(\d+.\d+)(?!.*\d)', x)[0])
new_df['Weight']
```

```python
new_df['GPU_name'] = df['Card màn hình'].apply(lambda x: 'On board' if 'Intel' in x else " ".join(x.split(" ")[:-1]))
new_df['GPU_name'].unique()
# drop 7 nhan CPU, 8 nhan CPU
```

```python
new_df['GPU_ram'] = df['Card màn hình'].apply(lambda x: '0' if 'Intel' in x else x.split(" ")[-1])
new_df['GPU_ram'].unique()
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(new_df)
```

```python
new_df = new_df[~new_df.CPU.str.contains("-GHz")]
new_df = new_df[~new_df.GPU_ram.str.contains("Graphics")]
new_df = new_df[~new_df.GPU_ram.str.contains("GPU")]
new_df['GPU_ram'] = new_df['GPU_ram'].apply(lambda x: re.findall(r'^\d+', x)[0])

with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(new_df)
```
