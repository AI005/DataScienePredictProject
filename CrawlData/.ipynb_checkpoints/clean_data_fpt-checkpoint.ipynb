```python
import pandas as pd
```

```python
df = pd.read_csv('fptdata_full.csv',index_col='STT')
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
columns = ["Company","CPU", "RAM", "Bus", "SSD", "HDD", "GPU_name", "GPU_ram", "Screen", "Độ phân giải","Weight" ,"Price"]
new_df = pd.DataFrame([], columns=columns)
new_df
```

```python
df['CPU'].unique()
```

```python
df = df[df['CPU'] != 'Apple M1']
df = df[df['Ổ cứng'] != 'Đang cập nhật']
```

```python
new_df['CPU'] = df['CPU']
import re
def get_kg(x):
    l = re.findall(r"[-+]?\d*\.\d+|\d+", x)
    return l[0]

new_df['Weight'] = df['Trọng lượng (kg)'].apply(get_kg)
```

```python
df['RAM'].unique()
```

```python
df['RAM'] = df['RAM'].str.split(',')
```

```python
new_df['RAM'] = df['RAM'].apply(lambda x: x[0].strip())
new_df['Bus'] = df['RAM'].apply(lambda x: x[2].strip() if len(x) >= 3 else None)
new_df['Bus'].fillna(new_df['Bus'].mode()[0], inplace=True)
```

```python
df['Ổ cứng'].unique()
```

```python
df['Ổ cứng'] = df['Ổ cứng'].str.split('&')
df['Ổ cứng']
```

```python
def get_amount_SSD(x):
    if(len(x) > 1):
        return x[1].strip().split(' ')[1]
    if 'SSD' in x[0]:
        return x[0].split(' ')[1]
    else:
        return 0
    
def get_amount_HDD(x):
    if(len(x) > 1):
        return x[0].strip().split(' ')[1]
    if 'HDD' in x[0]:
        return x[0].split(' ')[1]
    else:
        return 0

new_df['SSD']= df['Ổ cứng'].apply(get_amount_SSD)
new_df['HDD']= df['Ổ cứng'].apply(get_amount_HDD)
```

```python
print(new_df['SSD'].unique())
print(new_df['HDD'].unique())
```

```python
new_df['SSD'] = new_df['SSD'].replace('1',1000).astype(float)
new_df['HDD'] = new_df['HDD'].replace('1',1000).astype(float)
```

```python
df['Đồ họa'].unique()
```

```python
df['Đồ họa'] = df['Đồ họa'].str.split('&').apply(lambda x: x[0])
df['Đồ họa'].unique()
```

```python
def get_GPU_ram(x):
    if 'GB' not in x:
        return 0
    else:
        l = x.strip().split(' ')
        if l[-1] == 'GB':
            return l[-2]
        else:
            return l[-1][0:-2]
    
def get_GPU_name(x):
    if 'GB' not in x:
        return "On board"
    else:
        l = x.strip().split(' ')
        if l[-1] == 'GB':
            return ' '.join(l[:-2])
        else:
            return ' '.join(l[:-1])
        
        
new_df['GPU_name'] = df['Đồ họa'].apply(get_GPU_name)
new_df['GPU_ram'] = df['Đồ họa'].apply(get_GPU_ram)
new_df
```

```python
df['Màn hình'].unique()
```

```python
df['Màn hình'] = df['Màn hình'].str.split(',')
```

```python
new_df['Screen'] = df['Màn hình'].apply(lambda x: x[0][0:-1])
new_df['Độ phân giải'] = df['Màn hình'].apply(lambda x: x[1][:x[1].rfind(' ')])
new_df
```

```python
new_df['Price'] = df['Price'].apply(lambda x: float(x[0:-2].replace('.','', 5)))
new_df
```

Ta thống kê dữ liệu xem

```python
new_df['Company'] = df['Name'].apply(lambda x: x[0:x.find(' ')])
new_df
```

```python
new_df.to_csv('fix_data_fpt.csv', index=False)
```
