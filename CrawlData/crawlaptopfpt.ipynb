Vì trang của fpt shop là web động nên ta phải dùng selenium để crawl dữ liệu về.

```python
#imports here
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
import time

import pandas as pd
```

```python
url = "https://fptshop.com.vn/may-tinh-xach-tay"

#specify the path to chromedriver.exe (download and save on your computer)
driver = webdriver.Chrome('/usr/bin/chromedriver')
driver.implicitly_wait(10)
# wait = WebDriverWait(driver, 10)

#open the webpage
driver.get(url)
#Scroll to the end of the page
num_page = 1
while num_page < 9:
    num_page += 1
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5)#sleep_between_interactions
    not_now = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, '.btn-light '))).click()
```

```python
div_product_list = driver.find_elements_by_css_selector(".prd-lap")
len(div_product_list)
```

```python
columns=["Name", "CPU", "RAM", "Ổ cứng", "Card đồ họa", "Màn hình","Trọng lượng (kg)", "Price"]
df = pd.DataFrame([], columns=columns)
i = 0
for div_product in div_product_list:
    print(i)
    i += 1
    _dict = dict()
    _dict['Name'] = div_product.find_element_by_css_selector('.cdt-product__name').text
    try:
        _dict['Price'] = div_product.find_element_by_class_name('progress').text
    except:
        _dict['Price'] = div_product.find_element_by_class_name('price').text
    infos = div_product.find_element_by_class_name("cdt-product__config__param").find_elements_by_tag_name('span')
    for info in infos:
        _dict[info.get_attribute('data-title')] = info.get_attribute('innerText')
    df = df.append(_dict, ignore_index=True)
    
df
```

```python
df.to_csv('data_fptshop.csv')
```

```python
urls = []
for div_product in div_product_list:
    url = div_product.find_elements_by_tag_name('a')[0].get_attribute('href')
    urls.append(url)
urls
```

```python
columns=["CPU", "RAM", "Ổ cứng", "Đồ họa", "Màn hình","Trọng lượng (kg)", "Hệ điều hành"]
df_2 = pd.DataFrame([], columns=columns)
i = 0
while i < len(urls):
    url = urls[i]
    print(i)
    _dict = dict()
    if i == 38 or i == 200:
        df_2 = df_2.append(_dict, ignore_index=True)
        i += 1
        continue
        
    driver.get(url)
# #     try:
#     try:
#         driver.find_element_by_class_name('st-stt__noti')
#         i += 1
#         continue
#     except:
#         _dict["Price"] = driver.find_element_by_class_name('st-price').text
#     _dict["Name"] = driver.find_element_by_class_name('st-name').text
      
    list_info = driver.find_element_by_tag_name('table').text.split('\n')
    _dict["CPU"] = list_info[0].replace('CPU ', '')
    _dict["RAM"]  = list_info[1].replace('RAM ', '')
    _dict["Màn hình"] = list_info[2].replace("Màn hình ", '')
    _dict['Đồ họa'] = list_info[3].replace("Đồ họa ", '')
    _dict['Ổ cứng'] = list_info[4].replace("Ổ cứng ", '')
    _dict['Hệ điều hành'] = list_info[5].replace("Hệ điều hành ", '')
    _dict['Trọng lượng (kg)'] = list_info[6].replace("Trọng lượng (kg) ", '')
        
    df_2 = df_2.append(_dict, ignore_index=True)
    i += 1
#     except:
#         i = i
    
    
df_2
```
