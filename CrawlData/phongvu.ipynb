```python
#imports here
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
import time

import pandas as pd
```

```python
#specify the path to chromedriver.exe (download and save on your computer)
driver = webdriver.Chrome('/usr/bin/chromedriver')
driver.implicitly_wait(10)
wait = WebDriverWait(driver, 10)
```

```python
columns =['CPU', 'Chip đồ họa', 'RAM','Màn hình','Lưu trữ','Kích thước','Khối lượng']
df = pd.DataFrame([], columns=columns)
i = 1
for url in  open('phongvu_links.txt', 'r'):
    print(i)
    i += 1
    driver.get(url)
    while True:
        try:
            wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, '.css-h95iyj'))).click()
            break;
        except:
            pass
    
    infos = driver.find_elements_by_class_name('css-7j9rw7') 
    _dict = dict()    
    for info in infos:
        l = info.text.split('\n')
        if len(l) == 2:
            if l[0] in columns:
                _dict[l[0]] = l[1]
        elif len(l) == 1:
            if l[0] in columns:
                _dict[l[0]] = ''
                
    df = df.append(_dict, ignore_index=True)
        
df
        
```
